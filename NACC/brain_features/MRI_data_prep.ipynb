# Rough Estimate for diagnostic value

```python
diag = []
for i in range(len(df_mri)):
    ID = df_mri.iloc[i]['NACCID']
    diag.append(round(df_uds[df_uds['NACCID']== ID]['NACCUDSD'].mean()))
 
```

```python
smri = df_mri.loc[:,'NACCICV':'RTRTEMM']
smri['NACCID']= df_mri['NACCID']
smri['NACCUDSD'] = diag
smri = smri[(smri['NACCUDSD']==1)| (smri['NACCUDSD']==2) | (smri['NACCUDSD']==3)| (smri['NACCUDSD']==4)]
smri
```

```python
summary = smri.describe()
summary
```

After close inspection of the max values for each feature it is found that only three columns have missing values

'NACCICV':9999.999 

'NACCWMVL':9999.999 

'WMHVOL':999.9999 

Now let's replace these fields with np.nan so that our algorithms later will deal with them in standard way

```python
smri_final = smri.copy()
smri_final.loc[(smri_final['NACCICV']== 9999.999) | (smri_final['NACCICV']== 8888.888),'NACCICV']= np.nan
smri_final.loc[(smri_final['NACCWMVL']== 9999.999) | (smri_final['NACCWMVL']== 8888.888),'NACCWMVL']= np.nan
smri_final.loc[(smri_final['WMHVOL']== 999.9999) | (smri_final['WMHVOL']== 888.8888),'WMHVOL']= np.nan
```

```python
smri_final.describe()
```

```python
smri_final
```

# Replace NaN with mean for that column

```python
smri_final[['NACCICV']] = smri_final[['NACCICV']].fillna(value=smri_final.NACCICV.mean())
smri_final[['NACCWMVL']] = smri_final[['NACCWMVL']].fillna(value=smri_final.NACCWMVL.mean())
smri_final[['WMHVOL']] = smri_final[['WMHVOL']].fillna(value=smri_final.NACCICV.mean())
```
