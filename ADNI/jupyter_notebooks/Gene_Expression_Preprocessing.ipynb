Previously, gene expression features in ADNI database was used as it is. After coming across the paper "https://www.nature.com/articles/s41598-020-60595-1" it felt necessary to perform following pre-processing before using the data for our use. 
1. Only keep the high quality gene expression data for subjects with RIN (RNA integrity number) value is >6.9 
2. To reduce the background noise of ANDI, exclude the probes with an intensity value ≤ the median of all gene expression values in 100 or more samples.
3. If there were multiple probes annotated in one gene, then the median value of those was selected.

```python
X = df.loc[:,'11715100_at':'AFFX-r2-TagQ-5_at']
X
```

```python
overall_median = X.iloc[1:].stack().median()
overall_median
```

```python
filt_cols = []
probes = X.columns
for probe in probes:
    expr = X[probe][1:].astype(float)
    if sum(expr > overall_median) >= 100:
        filt_cols.append(probe)
```

# T-test on the preprocessed gene expression data (11860 genes)

```python
df_CN = gene_df[gene_df['DX']=='CN']
df_AD = gene_df[gene_df['DX']=='Dementia']
df_MCI = gene_df[gene_df['DX']=='MCI']
```

```python
df_CN_exp = df_CN.loc[:,'RBMX2': 'DSN1'].astype('float')
df_AD_exp = df_AD.loc[:,'RBMX2': 'DSN1'].astype('float')
df_MCI_exp = df_MCI.loc[:,'RBMX2': 'DSN1'].astype('float')
```

```python
data = df_AD_exp['RBMX2'] #Normality test for just one example gene to be able to use student t test
from scipy import stats
stats.normaltest(data)
# null hypothesis: x comes from a normal distribution
```

```python
from scipy.stats import ttest_ind
genes = df_AD_exp.columns
groups = ['AD_CN', 'AD_MCI', 'CN_MCI']
df_p = pd.DataFrame(columns=groups, index = genes) #datafrane for collecting p values for test
for gene in genes:
    data1 = df_AD_exp[gene]
    data2 = df_CN_exp[gene]
    stat, p = ttest_ind(data1, data2)
    df_p.loc[gene,'AD_CN'] = stat,p

for gene in genes:
    data1 = df_AD_exp[gene]
    data2 = df_MCI_exp[gene]
    stat, p = ttest_ind(data1, data2)
    df_p.loc[gene,'AD_MCI'] = stat, p
    
for gene in genes:
    data1 = df_CN_exp[gene]
    data2 = df_MCI_exp[gene]
    stat, p = ttest_ind(data1, data2)
    df_p.loc[gene,'CN_MCI'] = stat, p

    
```
